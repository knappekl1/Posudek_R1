```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from decimal import *
from datetime import datetime
```

```python
df=pd.read_csv("../Phillip-Morris_adj.csv")
```

```python
df.head(20)
```

```python
df["Datum"]=pd.to_datetime(df["Datum"],format="%d.%m.%Y")
```

```python
#df["Datum"][0] + pd.Timedelta(days=1)
```

```python
df.sort_values("Datum",axis=0, inplace = True)
```

```python
i = len(df)
```

```python
i
```

```python
df["Objem BCPP[Kč]"] = df["Objem BCPP[Kč]"].str.replace("\s+","").apply(lambda x: Decimal(x))
df["Počet BCPP[ks]"] = df["Počet BCPP[ks]"].str.replace("\s+","").apply(lambda x: Decimal(x))
df["Kurz BCPP[Kč]"] = df["Kurz BCPP[Kč]"].str.replace("\s+","").apply(lambda x: Decimal(x))
```

```python
df_small = df[df["Datum"].between("2007-05-22", "2007-06-22")]
```

```python
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df_small.reset_index(drop=True, inplace=True)
```

```python
ind = np.arange(0,i,5)
```

```python
ind
```

```python
def GetDate(dataFrame, index_no):
    r = dataFrame["Datum"][index_no].strftime("%d.%m.%Y")
    return r
```

```python
labels = []
for  i in ind:
    labels.append(GetDate(df,i))
```

```python
ind_small = np.linspace(0,len(df_small)-1,num=5)
```

```python
ind_small = ind_small//1
```

```python
labels_small = []
for  i in ind_small:
    labels_small.append(GetDate(df_small,i//1))
```

```python
labels_small
```

```python
df_small["Datum"][23]
```

```python
fig, ax = plt.subplots(figsize=(18,10),dpi=150)
ax.set_ylim(6000,14000)
ax.plot(df.index,df["Kurz BCPP[Kč]"])
plt.xticks(ind,labels) #(position, label)
fig.autofmt_xdate() #auto format labels
ax2 = ax.twinx()
ax2.set_ylim(0,100000)
ax2.bar(df.index,df["Počet BCPP[ks]"], color="gray")
ax.set_xlabel("Datum")
ax.set_ylabel("Cena Kč")
ax2.set_ylabel("Objem ks")
plt.title("Kurs akcie Phillip-Morris na BCPP",fontsize=20)
ax3 = fig.add_axes([0.64,0.60,0.25, 0.25])
ax3.plot(df_small.index, df_small["Kurz BCPP[Kč]"], color="green")
plt.xticks(ind_small,labels_small) #(position, label)
fig.autofmt_xdate() #auto format labels
ax3.set_ylabel("Cena Kč")
ax3.set_title("Kurs 30denní")
plt.setp(ax3.get_xticklabels(), rotation=30, horizontalalignment='right')
#plt.tight_layout()
plt.show()
fig.savefig("P-M_Chart.png", bbox_inches="tight")
```
